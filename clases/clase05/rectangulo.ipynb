```python
#Inicializacion de las variables

ancho = 20
alto = 40

anchoValido = False
altoValido = False 
tipoValido = False
tipo=''
```

```python
print ('Bienvenido al programa que dibuja rectángulos')

#Valido y guardo el ancho.
while (altoValido==False) :
    valor= input('Establezca el ancho del rectángulo, <30:')
    #Valido que sea numerico y menor a 30
    if (str.isdigit(valor)):  # Funcion para validar si el texto es numero
        if (int(valor)< 30):
            alto=int(valor)
            altoValido=True
        else:
            print ('Por favor ingresa un valor < 30')
    else :
        print ('Debe ser tipo entero')
 

#Valido y guardo el alto.
while (anchoValido==False) :
    valor= input('Establezca el alto del rectangulo , <40:')
    #Valido que sea numerico y menor a 40
    if (str.isdigit(valor)) :  # Funcion para validar si el texto es numero
        if (int(valor)< 40) :
            ancho=int(valor)
            anchoValido=True
        else:
            print ('Por favor ingresa un valor < 40')
    else :
        print ('Debe ser tipo entero')

#Valido y guardo el marcador
while (tipoValido==False) :
    tipo= input('Que marcador usará, # o - o *, por favor seleccione una opción :')
    if (tipo=='#' or tipo=='*' or tipo=='-') :
        tipoValido=True
    else :
        print ('Marcador incorrecto, por favor vuelva a ingresarlo')
     
        
#Establezco las variables 
base= tipo * ancho
lados= tipo + ' ' *(ancho -2) + tipo 

#Dibujo

i=1

while i<=alto :
    if (i==1 or i==alto) :  #Dibujo lados del cuadradola base del rectangulo
        print (base)
    else:
        print(lados) #Dibujo lados del cuadrado
    i=i+1    

```
